```python
import numpy as np
test = align_obj.rrna

## identify gaps
gaps = []
for seq in test:
    gap_start = None
    for i, char in enumerate(seq):
        if char == '-' and gap_start is None:
            gap_start = i
        elif char != '-' and gap_start is not None:
            if i - gap_start > 50:
                gaps.append((gap_start, i-1))
            gap_start = None
    if gap_start is not None and len(seq) - gap_start > 50:
        gaps.append((gap_start, len(seq)-1))
# Sort gaps by start position
gaps.sort(key=lambda x: x[0])
# Group overlapping gaps
overlapping_groups = []
current_group = [gaps[0]]

for i in range(1, len(gaps)):
    current_gap = gaps[i]
    last_gap = current_group[-1]
    # Check if current gap overlaps with last gap in current group
    if current_gap[0] <= last_gap[1]:
        current_group.append(current_gap)
    else:
        overlapping_groups.append(current_group)
        current_group = [current_gap]

# Add the last group
if current_group:
    overlapping_groups.append(current_group)

# For each overlapping group, find the shortest gap that covers all gaps
final_gaps = []
for group in overlapping_groups:
    if len(group) == 1:
        final_gaps.append(group[0])
    else:
        # Find the shortest gap that ensures uncovered regions are <50 bases
        min_start = min(g[0] for g in group)
        max_end = max(g[1] for g in group)
        
        # Check all possible gaps within the range
        best_gap = None
        min_length = float('inf')
        
        for start in range(min_start, max_end + 1):
            for end in range(start, max_end + 1):
                # Check if this gap covers all gaps with uncovered regions <50
                valid = True
                for gap in group:
                    if gap[0] < start and start - gap[0] >= 50:
                        valid = False
                        break
                    if gap[1] > end and gap[1] - end >= 50:
                        valid = False
                        break
                
                if valid and (end - start + 1) < min_length:
                    min_length = end - start + 1
                    best_gap = (start, end)
        
        if best_gap:
            final_gaps.append(best_gap)

gaps = final_gaps

slice.nt_array
```
